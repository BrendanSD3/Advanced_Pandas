We will use the price column to see how expensive the unclean airBnB listings are

Write a function to convert the price column into a 
series of floats<br>

Use the <i>apply</i> function and your function to reformat the price column


=================================================

```python
def convertDollarStringToInt(dollar_string):
    dollar_str = str(dollar_string).strip('$').replace(',', '')
    return float(dollar_str)
```

```python
airBCalNums = airB_cal.price.apply(convertDollarStringToInt)
airB_cal.price = airBCalNums
airB_cal.set_index("listing_id")
```

===============================================

We want to look at the price of listings.
Some prices are NaNs - deal with themaccordingly!

===============================================


```python
airB_calNotNull= airB_cal.loc[airB_cal.price.notnull()]
airB_calNotNull
```

===============================================

There are prices for each listing over multiple days.

Get a figure for each listing of the average shown for every day.<br>

```python
avg_listing_price = airB_calNotNull.groupby("listing_id").price.mean().round(0)
avg_listing_price.mean()
```
